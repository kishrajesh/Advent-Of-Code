```python
puzzle = open("AoC18.txt").read().splitlines()
```

```python
lastSound = 0
registers = {}
i = 0
while(0<=i<len(puzzle)):
    p = puzzle[i].split()
    if(p[1] not in registers):
        registers[p[1]] = 0
    if(p[0]=="snd"):
        lastSound = registers[p[1]] if p[1] in registers else int(p[1])
    elif(p[0]=="set"):
        registers[p[1]] = registers[p[2]] if p[2] in registers else int(p[2])
    elif(p[0]=="add"):
        registers[p[1]] += registers[p[2]] if p[2] in registers else int(p[2])
    elif(p[0]=="mul"):
        registers[p[1]] *= registers[p[2]] if p[2] in registers else int(p[2])
    elif(p[0]=="mod"):
        registers[p[1]] %= registers[p[2]] if p[2] in registers else int(p[2])
    elif(p[0]=="rcv"):
        if(registers[p[1]] if p[1] in registers else int(p[1])!=0):
            print(lastSound)
            break
    elif(p[0]=="jgz"):
        if(registers[p[1]] if p[1] in registers else int(p[1])>0):
            i += registers[p[2]] if p[2] in registers else int(p[2])
            continue
    i += 1
```
