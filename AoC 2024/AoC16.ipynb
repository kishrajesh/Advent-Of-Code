```python
puzzle = open("AoC16.txt").read().splitlines()
puzzle = [list(x) for x in puzzle]
```

```python
nodes = set()
neighs = {}
for i in range(len(puzzle)):
    for j in range(len(puzzle[0])):
        if puzzle[i][j] in ".SE":
            for dir in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                nodes.add((i, j, dir))
#create edges between nodes to account for turning
for i in range(len(puzzle)):
    for j in range(len(puzzle[0])):
        if puzzle[i][j] in ".SE":
            for dir in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                for dir2 in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                    if dir == dir2:
                        continue
                    if (dir[0]+dir2[0], dir[1]+dir2[1]) == (0, 0):
                        continue
                    if (i+dir2[0], j+dir2[1], dir2) in nodes:
                        if (i, j, dir) not in neighs:
                            neighs[(i, j, dir)] = []
                        neighs[(i, j, dir)].append(((i, j, dir2), 1000))

for i in range(len(puzzle)):
    for j in range(len(puzzle[0])):
        if puzzle[i][j] in ".SE":
            neigh = []
            for dir in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                if (i+dir[0], j+dir[1], dir) in nodes:
                    neigh.append((i+dir[0], j+dir[1], dir))
            for dir in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                if (i, j, dir) not in neighs:
                    neighs[(i, j, dir)] = []
                for n in neigh:
                    old_dir = n[2]
                    s = (old_dir[0]+dir[0],old_dir[1]+dir[1])
                    if s == (0,0):
                        continue
                    if max(s)==2 or min(s)==-2:
                        neighs[(i, j, dir)].append((n, 1))
                                  
```

```python
S = ()
E = ()
for i in range(len(puzzle)):
    for j in range(len(puzzle[0])):
        if puzzle[i][j] == "S":
            S = (i, j)
        if puzzle[i][j] == "E":
            E = (i, j)
nodes.add((E[0], E[1], (0,0)))
#add directed edges from neighbors to E
neighs[(E[0], E[1], (0,0))] = []
for dir in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
    if (E[0], E[1], dir) in nodes:
        neighs[(E[0], E[1], dir)].append(((E[0], E[1], (0,0)), 0.1))
```

```python
#use dijkstras to find shortest path from S to E
#start node is (S[0], S[1])
#end node is (E[0], E[1])
#neighs is a dictionary of neighbors with weights
#nodes is a set of all nodes
dist = {}
for node in nodes:
    dist[node] = 100000000000000
dist[(S[0], S[1], (0,1))] = 0
unvisited = set(nodes)
from tqdm import tqdm
ctr = 0
while len(unvisited) > 0:
    ctr+=1
    #if ctr%100 == 0:
    #    print(len(unvisited))
    min_node = None
    for node in unvisited:
        if min_node == None or dist[node] < dist[min_node]:
            min_node = node
    unvisited.remove(min_node)
    for neigh in neighs[min_node]:
        if dist[min_node] + neigh[1] < dist[neigh[0]]:
            dist[neigh[0]] = dist[min_node] + neigh[1]
ans1 = dist[(E[0], E[1], (0,0))]-0.1
print(ans1)
```

```python
#determine which nodes are on a shortest path
#there may be more than one shortest path
dist = {}
preds = {}
for node in nodes:
    dist[node] = 100000000000000
    preds[node] = []
dist[(S[0], S[1], (0,1))] = 0
unvisited = set(nodes)
from tqdm import tqdm
ctr = 0
while len(unvisited) > 0:
    ctr+=1
    if ctr%100 == 0:
        print(len(unvisited))
    min_node = None
    for node in unvisited:
        if min_node == None or dist[node] < dist[min_node]:
            min_node = node
    unvisited.remove(min_node)
    for neigh in neighs[min_node]:
        if dist[min_node] + neigh[1] == dist[neigh[0]]:
            preds[neigh[0]].append(min_node)
        if dist[min_node] + neigh[1] < dist[neigh[0]]:
            dist[neigh[0]] = dist[min_node] + neigh[1]
            preds[neigh[0]] = []
            preds[neigh[0]].append(min_node)
        
def get_visited(node):
    visited = set()
    queue = [node]
    while len(queue) > 0:
        n = queue.pop()
        visited.add(n)
        for pred in preds[n]:
            if pred not in visited:
                queue.append(pred)
    return visited
#find the number of nodes that are on at least one path
#including S and E
poss_nodes = set()
for path in preds[(E[0], E[1], (0,0))]:
    visited = get_visited(path)
    poss_nodes = poss_nodes.union(visited)
poss_positions = set()
for node in poss_nodes:
    poss_positions.add((node[0], node[1]))
ans2 = len(poss_positions)
print(ans2)
```
