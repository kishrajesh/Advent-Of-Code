```python
puzzle = list(open('AoC24.txt','r').read().splitlines())
for i in range(len(puzzle)):
    x=puzzle[i]
    ctr=0
    ans=[]
    while(ctr<len(x)):
        if(x[ctr]=='e' or x[ctr]=='w'):
            ans.append(x[ctr])
        else:
            p1=x[ctr]
            p2=x[ctr+1]
            ans.append(p1+p2)
            ctr+=1
        ctr+=1
    puzzle[i]=ans
print(puzzle)
```

```python
def nex(poin, mov):
    ctr=poin.copy()
    if(mov=='e'):
        ctr[0]=ctr[0]+2
    if(mov=='w'):
        ctr[0]=ctr[0]-2
    if(mov[0]=='n'):
        ctr[1]=ctr[1]+1
    if(mov[0]=='s'):
        ctr[1]=ctr[1]-1
    if(len(mov)==2):
        if(mov[1]=='e'):
            ctr[0]=ctr[0]+1
        if(mov[1]=='w'):
            ctr[0]=ctr[0]-1
    return ctr
```

```python
li=[] #list of shaded hexagons
for i in range(len(puzzle)):
    poin=[0,0]
    x=puzzle[i]
    for j in range(len(x)):
        poin=nex(poin, x[j])
    if(poin in li):
        li.remove(poin)
    else:
        li.append(poin)
print(len(li))
print(li)
```

```python
def neighs(poin):
    adj=[[2,0],[-2,0],[1,1],[-1,1],[-1,-1],[1,-1]]
    ans=[]
    for i in adj:
        x=poin.copy()
        x[0]+=i[0]
        x[1]+=i[1]
        ans.append(x)
    return ans
def test(poin, lis):
    x=neighs(poin)
    ctr=0
    for i in x:
        if i in lis:
            ctr+=1
    return ctr
```
