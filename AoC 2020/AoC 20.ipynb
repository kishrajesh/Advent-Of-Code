```python
def column(matrix, i):
    return [row[i] for row in matrix]
def compare(tilex,i,j):
    tile1=tilex[i]
    tile2=tilex[j]
    a=["","","",""]
    b=["","","",""]
    a[0]=tile1[0]
    a[2]=tile1[N-1]
    b[0]=tile2[0]
    b[2]=tile2[N-1]
    a[3]=column(tile1,0)
    a[1]=column(tile1,N-1)
    b[3]=column(tile2,0)
    b[1]=column(tile2,N-1)
    for x in range(len(a)):
        for y in range(len(b)):
            i=a[x]
            j=b[y]
            if(i==j):
                return x,True,False
            if(i[::-1]==j):
                return x,True,True
    return 0,False,False
```

```python
corners=[]
adja=[]
edges=[]
midd=[]
mape={}
side=[]
for i in range(len(tiles)):
    sides=[]
    adj=[]
    for j in range(len(tiles)):
        if(i!=j and (compare(tiles,i,j)[1])):
            if(compare(tiles,i,j)[0] not in sides):
                sides.append(compare(tiles,i,j)[0])
            adj.append(j)
            #adj.append(compare(tiles,i,j)[2])
    #print(sides, len(sides), adj)
            #print(compare(tiles,i,j)[0])
    if(len(sides)==2):
        corners.append(i)
    if(len(sides)==3):
        edges.append(i)
    if(len(sides)==4):
        midd.append(i)
    side.append(sides)
    adja.append(adj)
    mape[names[i]]=i
prod=1
for i in range(4):
    prod*=names[corners[i]]
print(prod)
```

```python
def rotateMatrix(mat, turn): 
    x=mat
    #if(type(x)==list):
     #   x=np.array(x)
    for i in range(turn):
        x=np.rot90(x)
    return x
```

```python
def combineO(tile1, tile2, tile3):
    a=["","","",""]
    b=["","","",""]
    c=["","","",""]
    a[0]=tile1[0]
    a[2]=tile1[N-1]
    b[0]=tile2[0]
    b[2]=tile2[N-1]
    c[0]=tile3[0]
    c[2]=tile3[N-1]
    a[3]=column(tile1,0)
    a[1]=column(tile1,N-1)
    b[3]=column(tile2,0)
    b[1]=column(tile2,N-1)
    c[3]=column(tile3,0)
    c[1]=column(tile3,N-1)
    R=0
    L=0
    fl=False
    for x in range(len(a)):
        for y in range(len(b)):
            i=a[x]
            j=b[y]
            if(i==j):
                R=x
                L=y
            if(i[::-1]==j):
                R=x
                L=y
                fl=True
    a1=rotateMatrix(tile1, R-1)
    b1=rotateMatrix(tile2, (L+1)%4)
    if(fl):
        b1=np.fliplr(b1)
    B=a1[N-1]
    T=a1[0]
    ctr=False
    for y in range(len(c)):
        j=c[y]
        if(type(B)!=list):
            B=B.tolist()
        #print(B,j,"debug")
        if((B==j)):
            U=y
            ctr=True
        if(B==j[::-1]):
            U=y
            fl=True 
            ctr=True
        if(not ctr):
            if(type(T)!=list):
                T=T.tolist()
            if(T==j):
                U=y
            if(T==j[::-1]):
                U=y
                fl=True
    #print("c", c)
    if(ctr):
        c1=rotateMatrix(tile3,4-U)
        if(fl):
            c1=np.flipud(c1)
    else:
        c1=rotateMatrix(tile3,4-U)
        c1=np.flipud(c1)
        if(fl):
            c1=np.flipud(c1)        
    if(not ctr):
        a1=np.flipud(a1)
        b1=np.flipud(b1)
    b1=np.fliplr(b1) 
    #mute for test case
    c1=rotateMatrix(c1,2)
    a1=a1.tolist()
    b1=b1.tolist()
    c1=c1.tolist()
    return a1,b1,c1
```

```python
def fit(tile1, tile2, di):
    if(di=="LR"):
        end=column(tile1, N-1)
        for i in range(4):
            x=rotateMatrix(tile2,i)
            if(column(x,0)==end):
                return x
        y=np.fliplr(tile2)
        for i in range(4):
            x=rotateMatrix(y,i)
            if(column(x,0)==end):
                return x
    if(di=="UD"):
        end=tile1[N-1]
        for i in range(4):
            x=rotateMatrix(tile2,i)
            if(type(end)!=list):
                end=end.tolist()
            if(type(x)!=list):
                x=x.tolist()
            if(x[0]==end):
                return x
        y=np.flipud(tile2)
        for i in range(4):
            x=rotateMatrix(y,i)
            x=x.tolist()
            if(x[0]==end):
                return x
```
