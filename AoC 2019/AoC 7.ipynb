```python
puzzle = open("AoC7.txt", "r").read().splitlines()
puzzle = [int(i) for i in puzzle[0].split(",")]
```

```python
def paramModes(opcode):
    opcode = str(opcode)
    if len(opcode) < 3:
        return [0,0,0]
    elif len(opcode) == 3:
        return [int(opcode[0]),0,0]
    elif len(opcode) == 4:
        return [int(opcode[1]),int(opcode[0]),0]
    elif len(opcode) == 5:
        return [int(opcode[2]),int(opcode[1]),int(opcode[0])]
    return [0,0,0]
def intcode(puzz, inputs, st):
    #intcode computer
    p = puzz.copy()
    i = st
    ans = 0
    ctr = 0
    while i < len(p):
        if p[i] == 99:
            return None, None, None
            break
        elif p[i]%100 == 1:
            c,b,a = paramModes(p[i])
            p1,p2=p[i+1],p[i+2]
            if c == 0:
                p1 = p[p[i+1]]
            if b == 0:
                p2 = p[p[i+2]]
            p[p[i+3]] = p1 + p2
            i += 4
        elif p[i]%100 == 2:
            c,b,a = paramModes(p[i])
            p1,p2=p[i+1],p[i+2]
            if c == 0:
                p1 = p[p[i+1]]
            if b == 0:
                p2 = p[p[i+2]]
            p[p[i+3]] = p1 * p2
            i += 4
        elif p[i]%100 == 3:
            p[p[i+1]] = inputs[ctr]
            ctr += 1
            i += 2
        elif p[i]%100 == 4:
            c,b,a = paramModes(p[i])
            if c == 0:
                ans = (p[p[i+1]])
            else:
                ans = (p[i+1])
            return p, ans, i+2
            i += 2
        elif p[i]%100 == 5:
            c,b,a = paramModes(p[i])
            p1,p2=p[i+1],p[i+2]
            if c == 0:
                p1 = p[p[i+1]]
            if b == 0:
                p2 = p[p[i+2]]
            if p1 != 0:
                i = p2
            else:
                i += 3
        elif p[i]%100 == 6:
            c,b,a = paramModes(p[i])
            p1,p2=p[i+1],p[i+2]
            if c == 0:
                p1 = p[p[i+1]]
            if b == 0:
                p2 = p[p[i+2]]
            if p1 == 0:
                i = p2
            else:
                i += 3
        elif p[i]%100 == 7:
            c,b,a = paramModes(p[i])
            p1,p2=p[i+1],p[i+2]
            if c == 0:
                p1 = p[p[i+1]]
            if b == 0:
                p2 = p[p[i+2]]
            if p1 < p2:
                p[p[i+3]] = 1
            else:
                p[p[i+3]] = 0
            i += 4
        elif p[i]%100 == 8:
            c,b,a = paramModes(p[i])
            p1,p2=p[i+1],p[i+2]
            if c == 0:
                p1 = p[p[i+1]]
            if b == 0:
                p2 = p[p[i+2]]
            if p1 == p2:
                p[p[i+3]] = 1
            else:
                p[p[i+3]] = 0
            i += 4
        if(i>=len(p)):
            return None, None, None
            break
```

```python
import itertools
perms = list(itertools.permutations([0,1,2,3,4]))
def amplifier(puzzle, phase):
    ans = 0
    for i in phase:
        ans = intcode(puzzle, [i, ans], 0)[1]
    return ans
print(max([amplifier(puzzle, i) for i in perms]))
```
