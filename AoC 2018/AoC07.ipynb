```python
puzzle = open("AoC07.txt").read().splitlines()
puzzle = [p.split(" ") for p in puzzle]
puzzle = [[p[1], p[7]] for p in puzzle]
```

```python
ans1 = ""
deps = {}
for p in puzzle:
    if p[0] not in deps:
        deps[p[0]] = []
    if p[1] not in deps:
        deps[p[1]] = []
    deps[p[1]].append(p[0])
queue = []
for k in deps:
    if len(deps[k]) == 0:
        queue.append(k)
while len(queue) > 0:
    queue.sort()
    ans1 += queue[0]
    for k in deps:
        if queue[0] in deps[k]:
            deps[k].remove(queue[0])
    queue.remove(queue[0])
    for k in deps:
        if len(deps[k]) == 0 and k not in queue and k not in ans1:
            queue.append(k)
print(ans1)
```
