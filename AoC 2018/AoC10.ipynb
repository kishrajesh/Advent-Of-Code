```python
puzzle = open("AoC10.txt").read().splitlines()
puzzle = [p.split("<") for p in puzzle]
puzzle = [[p[1].split(">")[0], p[2].split(">")[0]] for p in puzzle]
positionx = [int(p[0].split(",")[0]) for p in puzzle]
positiony = [int(p[0].split(",")[1]) for p in puzzle]
velocityx = [int(p[1].split(",")[0]) for p in puzzle]
velocityy = [int(p[1].split(",")[1]) for p in puzzle]
```

```python
import matplotlib.pyplot as plt    
for j in range(10000):
    positionx = [positionx[i] + velocityx[i] for i in range(len(positionx))]
    positiony = [positiony[i] + velocityy[i] for i in range(len(positiony))]
    print(j)
    plt.figure(figsize=(15, 15))
    plt.scatter(positionx, positiony)
    plt.show()
```
