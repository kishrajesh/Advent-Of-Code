```python
puzzle = (open("AoC4.txt", "r").read().splitlines())
checksums = [i[-7:][1:6] for i in puzzle]
IDs = [int(i[i.rindex('-')+1:i.rindex('[')]) for i in puzzle]
names = [i[:i.rindex('-')] for i in puzzle]
print(puzzle)
```

```python
def maxi(s):
    freq=[]
    top=[]
    for i in range(26):
        x=s.count(chr(ord('a')+i))
        freq.append(x)
    while(len(top)<5):
        m=max(freq)
        for i in range(26):
            if(freq[i]==m):
                top.append(chr(ord('a')+i))
                freq[i]=0
    return ''.join(top)[:5]
```

```python
ans=0
good=[]
for i in range(len(puzzle)):
    if(maxi(names[i])==checksums[i]):
        ans+=IDs[i]
        good.append(i)
print(ans)
```

```python
def decode(s, ID):
    l=list(s)
    x=[]
    for i in range(len(s)):
        if(l[i]!="-"):
            x.append(chr((ord(l[i])-ord('a')+ID)%26+ord('a')))
        else:
            x.append('-')
    return(''.join(x))
```

```python
for i in good:
    print(decode(names[i], IDs[i]), IDs[i])
```
