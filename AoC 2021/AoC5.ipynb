```python
puzzle = (open("AoC5.txt", "r").read().splitlines())
puzzle=[i.split(" ") for i in puzzle]
puzzle=[[i[0].split(","),i[2].split(",")] for i in puzzle]
puzzle=[[[int(i[0][0]), int(i[0][1])],[int(i[1][0]), int(i[1][1])]] for i in puzzle]
```

```python
def line(vent):
    if(vent[0][0]==vent[1][0]):
        return True, vent[0][0], 0
    elif(vent[0][1]==vent[1][1]):
        return True, vent[0][1], 1
    return False,0
puzz2 = [i for i in puzzle if line(i)[0]]
```

```python
def has(vent, point):
    x,y,a,b=vent[0][0],vent[1][0],vent[0][1],vent[1][1]
    if(line(vent)[0]):
        if(min(x,y)<=point[0] and point[0]<=max(x,y) and min(a,b)<=point[1] and point[1]<=max(a,b)):
            return True
    else:
        if((x+a)==(y+b) and (x+a)==sum(point)):
            if(min(x,y)<=point[0] and point[0]<=max(x,y) and min(a,b)<=point[1] and point[1]<=max(a,b)):
                return True
        elif((x-a)==(y-b) and (x-a)==(point[0]-point[1])):
            if(min(x,y)<=point[0] and point[0]<=max(x,y) and min(a,b)<=point[1] and point[1]<=max(a,b)):
                return True
    return False
```

```python
miX=min(min([i[0][0] for i in puzz2]), min([i[1][0] for i in puzz2]))
maX=max(max([i[0][0] for i in puzz2]), max([i[1][0] for i in puzz2]))
miY=min(min([i[0][1] for i in puzz2]), min([i[1][1] for i in puzz2]))
maY=max(max([i[0][1] for i in puzz2]), max([i[1][1] for i in puzz2]))
```

```python
ans=0
for i in range(miX, maX+1):
    for j in range(miY, maY+1):
        ctr=0
        for k in puzz2:
            if(ctr==2):
                break
            elif(has(k, [i,j])):
                ctr+=1
        if(ctr==2):
            ans+=1
            print(i, ans)
print(ans)    
```
