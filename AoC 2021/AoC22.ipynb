```python
puzzle = (open("AoC22.txt", "r").read().splitlines())
bools=[i.split(" ")[0] for i in puzzle]
boxes=[i.split(" ")[1].split(",") for i in puzzle]
for i in range(len(boxes)):
    a,b,c=boxes[i]
    n=[a.split("=")[1].split(".."), b.split("=")[1].split(".."), c.split("=")[1].split("..")]
    n=[[int(i[0]), int(i[1])] for i in n]
    boxes[i]=n
ma=[max([i[0][1]] for i in boxes)[0], max([i[1][1]] for i in boxes)[0], max([i[2][1]] for i in boxes)[0]]
mi=[min([i[0][0]] for i in boxes)[0], min([i[1][0]] for i in boxes)[0], min([i[2][0]] for i in boxes)[0]]
print(ma)
print(mi)
from tqdm import tqdm
def cubes(mi, ma):
    #creates a 3D list of true or false values
    #d[a][b][c] denotes if cube in spot a,b,c is lighted
    #Initiates everything to False
    d=[[[False for k in range(mi[0], ma[0]+1)] for j in range(mi[1], ma[1]+1)] for i in range(mi[2], ma[2]+1)]
    return d
```

```python
def change(cubes, x, y, z, boo, xra, yra, zra):
    c=cubes.copy()
    if(boo=="on"):
        boo=True
    else:
        boo=False
    for i in tqdm(range(max(x[0], xra[0]), min(x[1], xra[1])+1)):
        for j in range(max(y[0], yra[0]), min(y[1], yra[1])+1):
            for k in range(max(z[0], zra[0]), min(z[1], zra[1])+1):
                if(xra[0]<=i<=xra[1] and yra[0]<=j<=yra[1] and zra[0]<=k<=zra[1]):
                    c[i][j][k]=boo
    return c
c=cubes([-50,-50,-50],[50,50,50])
```

```python
for i in range(len(bools)):
    c=change(c, boxes[i][0], boxes[i][1], boxes[i][2],bools[i],[-50,50],[-50,50],[-50,50])
```

```python
ctr=0
for i in c:
    for j in i:
        for k in j:
            if(k==True):
                ctr+=1
print(ctr)
```
